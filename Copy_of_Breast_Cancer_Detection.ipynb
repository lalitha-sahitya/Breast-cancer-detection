```python
import pandas as pd
import opendatasets as od
od.download('https://www.kaggle.com/datasets/yasserh/breast-cancer-dataset')
```

```python
os.listdir('breast-cancer-dataset')
```

```python
df=pd.read_csv('/content/breast-cancer-dataset/breast-cancer.csv')
df
```

```python
df.info()
```

```python
df=df.drop('id',axis=1)
```

```python
df['diagnosis']=df.diagnosis.map({'M':1,'B':0})
df
```

```python
df_new=df.corr()['diagnosis'].sort_values(ascending=False).head(25)
df_new
```

```python
column_names = df_new.index.tolist()
```

```python
df_new=df[column_names]
df_new
```

```python
x=df_new.drop('diagnosis',axis=1)
y=df_new['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X_train)
```

```python
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier,BaggingClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
def try_model(model):
  model.fit(X_train,y_train)
  train_preds=model.predict(X_train)
  test_preds=model.predict(X_test)
  train_rmse=accuracy_score(y_train,train_preds)
  test_rmse=accuracy_score(y_test,test_preds)
  return train_rmse,test_rmse
```

```python
dtc=DecisionTreeClassifier(random_state=42,max_features=0.9,max_depth=3,max_leaf_nodes=10)
try_model(dtc)
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
try_model(knn)
```

```python
lr=LogisticRegression()
try_model(lr)
```

```python
svmr=SVC(kernel='rbf')
try_model(svmr)
```

```python
svml=SVC(kernel='linear')
try_model(svml)
```

```python
bg=BaggingClassifier(n_estimators=20,random_state=42,n_jobs=-1)
try_model(bg)
```

```python
gb=GradientBoostingClassifier(n_estimators=20,random_state=42)
try_model(gb)
```

```python
rfc=RandomForestClassifier(n_estimators=100,random_state=42,n_jobs=-1)
try_model(rfc)
```

```python
ad=AdaBoostClassifier(n_estimators=100,random_state=42)
try_model(ad)
```

```python
test_preds=ad.predict(X_test)
test_preds
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
models=[lr, dtc, knn,
       svmr, svml, rfc, bg, ad, gb]
model_abbreviations = ['lr', 'dtc', 'knn', 'svmr', 'svml', 'rfc', 'bg', 'ad', 'gb']
```

```python
results=[]
for i, model in enumerate(models):
    metric_value = try_model(model)
    results.append((model_abbreviations[i], metric_value[1]))
```

```python
results_df = pd.DataFrame(results, columns=['models', 'Metric'])
sns.barplot(x='models', y='Metric', data=results_df)
plt.title('Model Performance')
plt.xlabel('Model')
plt.ylabel('Metric')
plt.xticks(rotation=45)
plt.show()
```
